```python
# Cleaning the Installs column
df["Installs"] = pd.to_numeric(df["Installs"].astype(str).str.replace(r"[+,]","",regex=True))
```

```python
df.Installs
```

```python
# Cleaning the Price Columns by convrting it to the Numeric data
df["Price"] = pd.to_numeric(df["Price"].str.replace("$",""))
```

```python
# Cleaning the Installs column by converting it to the numeric data
df["Installs"] = pd.to_numeric(df["Installs"].astype(str).str.replace(r"[+,]","",regex=True))
```

```python
# Cleaning the data by replacing the values with the numeric values

# Replaces the "Varies with the device" with the NaN
from pyparsing import Regex


df["Size"] = df["Size"].replace("Varies with device",np.nan)

# Convet the "K" values
mask_k = df["Size"].str.contains("k", na=False)
df.loc[mask_k,"Size"] = df.loc[mask_k,"Size"].str.replace("k","",regex=False).astype(float)*1000

mask_M = df["Size"].str.contains("M",na=False)
df.loc[mask_M,"Size"] = df.loc[mask_M,"Size"].str.replace("M","",regex=False).astype(float)*1000000

df["Size"]=pd.to_numeric(df["Size"],errors="coerce")
```
